## 1. Generalization - Section 4
Data prediction using our trained composite model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('./test_data.csv').drop(columns=["Index", " Net Income Flag"])

def preprocess_data(df):
    """Performs standardization and feature selection on training data set"""
    df_bankrupt = df['Bankrupt?']
    df_numpy = df.to_numpy()
    scalar = StandardScaler()
    X_scaled = scalar.fit_transform(df_numpy)

    df_scaled = pd.DataFrame(
        X_scaled,
        columns=df.columns,
        index=df.index
    )
    df_scaled['Bankrupt?'] = df_bankrupt #We reset bankrupt column since it does not need scaling

    corr_matrix = df_scaled.corr()

    bankrupt_feature = corr_matrix["Bankrupt?"]
    abs_corr = bankrupt_feature[abs(bankrupt_feature) > 0.1]
    features = df_scaled[abs_corr.index.tolist()]

    return features, corr_matrix

joblib.dump(preprocess_data, 'preprocess_data.joblib')
print("Custom preprocessor for training data 'preprocess_data' saved to 'preprocess_data.joblib'. Import to transform data.")
```
